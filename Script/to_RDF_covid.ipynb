```python
def to_float(series):
    series = series.str.replace(',','.').str.replace('−','-').astype(float)
    return series
```

```python
def to_float_coma(series):
    series =  series.str.replace(',','').astype(float)
    return series
```

```python
def to_float_dot(series):
    series =  series.str.replace('.','').astype(float)
    return series
```

```python
def line_prepender(filename, line):
    with open(filename, 'r+') as f:
        content = f.read()
        f.seek(0, 0)
        f.write(line.rstrip('\r\n') + '\n' + content)
```

```python
def add_list(filename, lines):
    lines.reverse()
    for i in lines:
        line_prepender(filename, i)
```

```python
def todate(str):
    year = str[:4]
    month = str[4:6]
    day = str[6:]
    return year + '-' + month + '-' + day
```

```python
# SOLO UNA FECHA, DAILY REPORT
for date in dates:
    #print(date)
    init = date
    end = date
    interval = []
    interval.append('c19-interval:Date' + date + ' a time:Interval ;\n')
    interval.append('    time:hasBeginning "' + todate(init) + '"^^xsd:date ;\n')
    interval.append('    time:hasEnd "' + todate(end) + '"^^xsd:date ;\n')
    interval.append('    rdfs:label "' + date + '" .\n')
    interval.append('\n\n')
    output = path_ttl + "interval" + file + ".ttl"
    file1 = open(output, "a")  # append mode 
    file1.writelines(interval) 
    file1.close()

file1 = open(output,"r+")  
  
print("Output of Read function is ")
print(file1.read())
```
